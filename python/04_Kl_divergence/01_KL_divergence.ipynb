```python
import numpy as np
from scipy.stats import norm
import pandas as pd
def kl_divergence(p, q):
    return np.sum(np.where(p != 0, p * np.log(p / q), 0))
kl_dict={}
given=[['02_2015_-_2018_Demographic_Snapshot_Pre-_K_For_All.csv','04_2013_-_2018_Demographic_Snapshot_District.csv'],['Total Enrollment','% Asian']]
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
for dataset in given[0]:
    data=pd.read_csv(dataset)
    numerical_vars = list(data.select_dtypes(include=numerics).columns)
    data = data[numerical_vars]
    att1_dict={}
    for att in given[1]:
        att2_dict={}
        for att2 in data.columns:
            att2_dict[att2]=kl_divergence(data[att],data[att2])
            #print(att," of ",dataset[:15]," has kl divergence of :",kl_divergence(data[att],data[att2])," with ",att2)
        att1_dict[att]=att2_dict
    kl_dict[dataset]=att1_dict 
kl_dict
            
```
